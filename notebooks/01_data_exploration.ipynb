```python
import ast

# Convert strings to actual lists
def parse_genres(genre_string):
    try:
        # ast.literal_eval converts string representation of a list to an actual list
        return ast.literal_eval(genre_string)
    except:
        return []

df['genres_list'] = df['genres'].apply(parse_genres)

print(df['genres_list'].head(10))
```

```python
from collections import Counter

all_genres = []
for genres_list in df['genres_list']:
    all_genres.extend(genres_list)

genre_counts = Counter(all_genres)

print("=== TOP 50 MOST COMMON GENRES ===")
for genre, count in genre_counts.most_common(50):
    print(f"{genre}: {count}")
```

```python
GENRE_GROUPS = {
    # 1. ROCK Y DERIVADOS
    'Rock': [
        'rock', 'classic rock', 'album rock', 'hard rock',
        'soft rock', 'modern rock', 'alternative rock', 'indie rock',
        'progressive rock', 'symphonic rock', 'folk rock', 'post-grunge'
    ],
    
    # 2. METAL
    'Metal': [
        'metal', 'alternative metal', 'nu metal', 'heavy metal',
        'death metal', 'black metal', 'metalcore'
    ],
    
    # 3. POP
    'Pop': [
        'pop', 'dance pop', 'pop dance', 'indie pop', 
        'art pop', 'electropop', 'synth pop', 'pop rock',
        'teen pop', 'power pop'
    ],
    
    # 4. HIP HOP / RAP
    'Hip Hop': [
        'hip hop', 'rap', 'pop rap', 'trap', 'conscious hip hop',
        'gangster rap', 'southern hip hop', 'east coast hip hop',
        'west coast rap', 'emo rap'
    ],
    
    # 5. ELECTRONIC / EDM
    'Electronic': [
        'edm', 'electro house', 'progressive house', 'house',
        'techno', 'trance', 'dubstep', 'drum and bass',
        'indietronica', 'electro'
    ],
    
    # 6. CLASSICAL
    'Classical': [
        'classical', 'baroque', 'classical era', 
        'late romantic era', 'early romantic era',
        'post-romantic era', 'early music', 
        'german romanticism', 'german baroque'
    ],
    
    # 7. JAZZ
    'Jazz': [
        'jazz', 'vocal jazz', 'smooth jazz', 'contemporary jazz',
        'bebop', 'cool jazz', 'jazz fusion'
    ],
    
    # 8. R&B / SOUL
    'R&B': [
        'r&b', 'soul', 'neo soul', 'contemporary r&b',
        'funk', 'motown'
    ],
    
    # 9. COUNTRY
    'Country': [
        'country', 'contemporary country', 'country road',
        'classic country', 'country rock', 'bluegrass'
    ],
    
    # 10. LATIN
    'Latin': [
        'latin pop', 'reggaeton', 'latin', 'salsa', 
        'bachata', 'regional mexican', 'banda'
    ],
    
    # 11. FOLK
    'Folk': [
        'folk', 'indie folk', 'folk rock', 'contemporary folk',
        'scottish folk', 'irish folk'
    ],
    
    # 12. ALTERNATIVE
    'Alternative': [
        'alternative', 'alternative rock', 'indie', 
        'permanent wave', 'new wave', 'post-punk'
    ],
    
    # 13. SINGER-SONGWRITER
    'Singer-Songwriter': [
        'singer-songwriter', 'lilith', 'mellow gold',
        'folk-pop'
    ],
    
    # 14. EASY LISTENING
    'Easy Listening': [
        'easy listening', 'adult standards', 'lounge',
        'bossa nova'
    ],
    
    # 15. SOUNDTRACK
    'Soundtrack': [
        'soundtrack', 'orchestral soundtrack', 'score',
        'video game music'
    ]
}
```

```python
def map_to_general_genre(genre_list):
    if not genre_list or genre_list == []:
        return None

    for general, specifics in GENRE_GROUPS.items():
        for genre in genre_list:
            if genre.lower() in specifics:
                return general
    return 'Other'

df['general_genre'] = df['genres_list'].apply(map_to_general_genre)

print("=== GENERAL GENRES DISTRIBUTION ===")
print(df['general_genre'].value_counts())
print(f"\nTotal tracks with genre: {df['general_genre'].notna().sum()}")
print(f"Tracks without mapping (Other): {(df['general_genre'] == 'Other').sum()}")
```

```python
# Ver qué porcentaje del dataset conservas

# Tracks con géneros mapeados
df_mapped = df[df['general_genre'].notna()].copy()
df_mapped = df_mapped[df_mapped['general_genre'] != 'Other']

# Tracks con audio features completos
audio_cols = ['tempo', 'energy', 'danceability', 'valence', 'acousticness',
            'instrumentalness', 'liveness', 'loudness', 'speechiness']
df_clean = df_mapped.dropna(subset=audio_cols)

print(f"Original dataset: {len(df):,} tracks")
print(f"With mapped genre: {len(df_mapped):,} tracks ({len(df_mapped)/len(df)*100:.1f}%)")
print(f"With genre + audio features: {len(df_clean):,} tracks ({len(df_clean)/len(df)*100:.1f}%)")
print(f"\nDiscarded tracks: {len(df) - len(df_clean):,}")
```

```python
print("=== FINAL DISTRIBUTION BY GENRE ===\n")
genre_distribution = df_clean['general_genre'].value_counts()
print(genre_distribution)
print(f"\nTotal genres: {len(genre_distribution)}")

# Estadísticas
print(f"\nMaximum tracks per genre: {genre_distribution.max():,}")
print(f"Minimum tracks per genre: {genre_distribution.min():,}")
print(f"Average tracks per genre: {genre_distribution.mean():.0f}")
print(f"Median tracks per genre: {genre_distribution.median():.0f}")
```

```python
sns.barplot(x=genre_distribution.index, y=genre_distribution.values, color='#7c6')
plt.xlabel('Genre')
plt.ylabel('Number of Tracks')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Parámetros
MIN_TRACKS = 10_000
MAX_TRACKS = 40_000

# Balancear dataset
df_balanced_list = []

print("=== BALANCING PROCESS ===\n")

for genre in df_clean['general_genre'].unique():
    genre_df = df_clean[df_clean['general_genre'] == genre]
    n_tracks = len(genre_df)

    if n_tracks < MIN_TRACKS:
        print(f"❌ {genre:20s}: {n_tracks:6,} tracks (< {MIN_TRACKS:,}) - DISCARDED")
        continue

    elif n_tracks > MAX_TRACKS:
        sampled = genre_df.sample(n=MAX_TRACKS, random_state=64)
        df_balanced_list.append(sampled)
        print(f"📉 {genre:20s}: {n_tracks:6,} → {MAX_TRACKS:6,} tracks (REDUCED)")

    else:
        df_balanced_list.append(genre_df)
        print(f"✅ {genre:20s}: {n_tracks:6,} tracks (MAINTAINED)")
# Concatenar
df_balanced = pd.concat(df_balanced_list, ignore_index=True)

# Shuffle para mezclar géneros
df_balanced = df_balanced.sample(frac=1, random_state=64).reset_index(drop=True)

print(f"\n{'='*60}")
print(f"🎯 Final Balanced Dataset")
print(f"{'='*60}")
print(f"Total tracks:  {len(df_balanced):,}")
print(f"Total genres: {df_balanced['general_genre'].nunique()}")
print(f"\nFinal distribution:")
print(df_balanced['general_genre'].value_counts().sort_values(ascending=False))
```

```python
# Seleccionar columnas finales
columns_final = [
    # Identificadores
    'track_id', 
    'name',
    
    # Audio Features (para embeddings numéricos)
    'acousticness', 
    'danceability', 
    'energy', 
    'instrumentalness', 
    'key',
    'liveness', 
    'loudness', 
    'speechiness', 
    'tempo', 
    'valence',
    
    # Géneros (para embeddings de género)
    'general_genre',
    'genres_list',
    
    # Metadata
    'popularity',
    'artist_popularity',
    'duration_ms',
    'explicit',
    'album_release_date'
]

df_final = df_balanced[columns_final].copy()

# Verificar nulos finales
print("\n=== QUALITY CHECK ===")
null_counts = df_final.isnull().sum()
if null_counts.sum() > 0:
    print("\nNull values found:")
    print(null_counts[null_counts > 0])
else:
    print("\n✅ No null values!")

# Info del dataset
print("\n=== DATASET INFO ===")
print(f"Shape: {df_final.shape}")
print(f"Memory: {df_final.memory_usage(deep=True).sum() / 1024**2:.1f} MB")

# Sample de cómo se ve
print("\n=== SAMPLE ===")
print(df_final.head())

# Guardar
from pathlib import Path
output_path = Path("../data/processed/spotify_clean_balanced.csv")
output_path.parent.mkdir(parents=True, exist_ok=True)
df_final.to_csv(output_path, index=False)

print(f"\n{'='*60}")
print(f"✅ DATASET SAVED SUCCESSFULLY")
print(f"{'='*60}")
print(f"Location: {output_path}")
print(f"Size: {output_path.stat().st_size / 1024 / 1024:.1f} MB")
print(f"Ready for embedding generation!")
```
